```python
def check_missing_values(data: pandas.DataFrame):
    return data[categorical].isnull().sum()
check_missing_values(data)
```

```python
from typing import List
def impute(data:pandas.DataFrame, list_of_columns: List[str]):
    for variable in list_of_columns: data[variable].fillna(data[variable].mode()[0], inplace=True)
    return check_missing_values(data)
```
